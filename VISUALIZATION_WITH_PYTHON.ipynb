DATA EXPLORATION WITH PYTHON

Types of Exploratory Data Analysis

UniVariate Analysis:

Histplot (String Type}

Countplot (string type)

BiVariate Analysis:

Barplot (String Type and Numerical Type Column)

Scatterplot (Both Numerical Type)

Correlation Matrix

Linear Regression

MultiVariate Analysis:

Barplot (String and Numerical Type and a HUE)

Correlation Matrix

Scatterplot (Both Numerical and a HUE)


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salary = pd.read_csv('/content/salaries_clean.csv')
```

```python
#count plot
plt.figure(figsize = (12, 6))
sns.countplot(x = salary.job_title_category)
```

```python
#distplot
sns.set(style = 'darkgrid')
sns.distplot(salary.annual_base_pay)
```

```python
plt.figure(figsize = (12, 6))
sns.histplot(x = salary.job_title_category, kde = True)
```

```python
salary.columns.unique()
```

```python
car = pd.read_csv('/content/car_prices.csv')
```

```python
sns.barplot(x = car.Condition, y = car.Mileage, ci = 0)
```

```python
sns.scatterplot( x = car['Engine Size'],  y = car.price,)
```

```python
sns.regplot(x = car['Year of manufacture'], y = car.price)
```

```python
sns.lineplot(x = car['Year of manufacture'], y = car.price, ci =0)
```

```python
sns.boxplot(car[['Year of manufacture', 'price', 'Engine Size']])
```

```python
sns.countplot(x = car['Condition'], hue = car['gear type'] )
```

```python
plt.figure(figsize =(12, 5))
sns.countplot(x = salary.job_title_category, hue = salary.job_title_rank)
```

```python
plt.figure(figsize = (6,3))
sns.heatmap(car[['Year of manufacture', 'price']].corr(), annot = True)
```

```python
sns.barplot(x = car.Condition, y = car.price, hue = car ['fuel type'],ci = 0)
```

```python
# -------------- Correlation Matrix For Multivariate -----------------
plt.figure(figsize = (12, 5))
sns.heatmap(car.corr(), annot = True, cmap = 'BuPu')
```

```python
# - In the US, what Software Job Title is the most paid.

salary.columns.unique()
```

```python
# - In the US, what Software Job Title is the most paid.
soft = salary.loc[salary.location_country == 'US']
soft = soft.loc[soft.job_title_category != 'Other']
plt.figure(figsize = (12, 6))
sns.set(style = 'darkgrid')
sns.barplot(x = soft.job_title_category, y = soft.annual_base_pay, ci = 0)
```

```python
# - Show the distribution of Annual Base Pay. Is it Normally Distributed.
sns.distplot(salary.annual_base_pay, kde = True)
```

```python
# - In Canada, what state has the highest paid Software Engineers.
Soft_CA = salary.loc[salary.location_country == 'CA']
sns.barplot(x = Soft_CA.location_state, y = Soft_CA.annual_base_pay, ci = 0)
```

```python
# - Choose 3 countries apart from the US, show what country pays Management Workers the most

Management = salary.loc[salary.job_title_category == 'Management']
Management = Management.loc[(Management.location_country == 'ES')| (Management.location_country == 'CA')|(Management.location_country == 'ZA')]

sns.barplot(x = Management.location_country, y = Management.annual_base_pay, ci = 0 )
```

```python
# - Using a barplot, show the difference between the Annual Bonus of Management Workers and Applied Science
diff = salary.loc[(salary.job_title_category == 'Management') | (salary.job_title_category == 'Applied Science')]
sns.barplot(x = diff.job_title_category, y = salary.annual_base_pay, ci = 0)

```
